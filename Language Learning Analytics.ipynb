##4.Algorithm
1.Import the required library.
2.Open the CSV file.
3.Read the data line by line.
4.Store each row into a list.
5.Count vocabulary for each language.
6.Count vocabulary for each level.
7.Display the results.
8.Draw a bar chart.
9.Draw a pie chart.

```python
import matplotlib.pyplot as plt
```

```python
file=open("Language Learning Analytics.csv","r",encoding="utf-8-sig")
lines=file.readlines()
file.close()
print(lines)
```

```python
data=[]

for line in lines[1:]:
    line=line.strip()
    row=line.split(",")
    data.append(row)
```

```python
for row in data:
    print(row)
```

```python
language_count={}

for row in data:
    language=row[0]

    if language in language_count:
        language_count[language]+=1
    else:
        language_count[language]=1
print("Vocabulary Count by Language")
print(language_count)
```

```python
level_count={}

for row in data:
    level=row[1]

    if level in level_count:
        level_count[level]+=1
    else:
        level_count[level]=1
print("Vocabulary Count by Level")
print(level_count)
```

```python
languages=list(language_count.keys())
counts=list(language_count.values())

plt.figure(figsize=(6,4))
plt.bar(languages,counts)
plt.title("Vocabulary Count by Language")
plt.xlabel("Language")
plt.ylabel("Number of Words")
plt.show()
```

```python
levels=list(level_count.keys())
counts=list(level_count.values())

plt.figure(figsize=(6,6))
plt.pie(counts, labels=levels, autopct="%1.1f%%", startangle=90)
plt.title("Vocabulary Distribution by Level")
plt.show()
```

##5.Conclusion
This analysis shows the distribution of vocabulary among French, Japanese, and Spanish. It also illustrates the proportion of vocabulary at different proficiency levels. This project demonstrates how Python can be used to organize, analyze, and visualize language learning data.
